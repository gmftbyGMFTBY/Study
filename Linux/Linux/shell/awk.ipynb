# AWK - 文本报告生成器
AWK是Unix和Linux下一款非常强大的文本报告生成器，可以帮助我们快速的实现对一个文本的格式化和读取　　

可以实现的目的 : 
    * 定义变量保存数据
    * 使用算数或者字符串运算操作数据
    * 引入结构化编程概念
    * 提取数据元素，生成可视化报告

## 基础
命令：我们的AWK的命令基本上都是这样的结构　　 
```
awk [options] 'Program' file1 file2 file3 ...
- 'Program':
pattern + action
```
1. 其中的program中主要使用两种语句，分别叫做模式（条件）和动作
2. 其中的options是我们的awk的参数，参数主要有
    * -v :指定变量，之后我们会讲 
    * -f :指定awk脚本调用
    * -F :指定我们的分隔符号
3. 我们需要认识到，awk如果没有在修改默认的参数的时候，是一行的一行的读取文件并且处理（以行为单位）这里默认使用换行符来表示行，然后在行内将我们的文本按照域分隔符切割成不同的字段
4. 在AWK中，我们的所有的\$开头的数字变量都代表我们的某一行的域，如果要使用变量，我们直接使用不加\$(这里和shell的语法进行区分)，加上了\$后面的跟的数字都会被我们整体的转义理解成是某一行的一个字段
5. 连续的分隔符我们都当成一个分隔符处理
6. 
```
awk '{print $1,$2}' test
```
这里面我需要强调几个语法点
    * print之后的都好表示的是使用默认的输出分隔符连接两个变量，如果没有逗号，我们默认是直接拼接 
    * 如果我们的行上没有\$1,\$2缺少那个我们就不输出哪个,\$0默认输出全行，不加参数默认输出全行
    * print之后可以跟上我们的默认的字符串进行输出
    * 对于我们的program每部的内容，我们需要了解：禁止使用单引号，双引号慎用，因为"\$1"会直接输出\$1并不会实现变量的替代

## 分割符
1. 分隔符分为输入分隔符和输出分隔符 - FS / OFS
2. 分隔符就是我们的在读入每一行之后按照什么样的字符去分割一个行成多个字段，默认的输出和输入的分隔符都是空格（默认空格来分割，默认空格来拼接输出 - 逗号） 
3. 重新指定输入分割符
    * -F'[xxx]':xxx是多种分隔符，我们可以制定多种符号都作为分隔符
    * -v FS="X":利用内置的变量和-v参数来修改我们的输入或者输出的分隔符（FS / OFS）
    ```
    awk -v FS=":" -v OFS="\t" '{print $0}'　file1 ...
    ```
    
## AWK变量
1.AWK的变量分为内置变量和自定义变量两种  
2.内置变量　　
    * FS : 输入分隔符，这里我们可以修改我们的分割符
    * OFS : 输出分隔符，修改我们的空格分隔符为自己指定的任意字符
    * RS : 记录输入分隔符，默认使用'\n'作为分隔符，这里可以自己指定我们的awk使用什么字符或者字符串来作为我们的一行的标准
    * ORS : 记录输出分隔符，默认的使用'\n'来作为一行的结束，这里采用自己制定任意字符或者字符串来作为我们的记录的换行符号
    * NR : 处理行数，当前处理的总行数，因为可以处理多个文本，多个文本的函数是累加的
    * FNR : 按文件来计算处理行数，按照每一个文件为单位来处理我们的行数
    * NF : 某一行的字段的总数，某一个行的分割后的字段的个数
    * FILENAME : 当前处理的文件名
    * ARGC : 命令行参数个数，首先注意我们的处理的文件名也是我们的命令行参数之一
    * ARGV : 命令行参数数组，0号是字符串'awk'，之后的是我们的处理文本的名称，也可以字节最后制定我们的命令行参数
3.自定义变量  
    1. 实现方式：
        * -v指定变量（变量名称区分大小写）
        * 在program语句中进行定义，注意多个语句我们采用分好分隔语句
    2. 注意点，我们允许在-v制定变量的时候，引用我们的ｓｈｅｌｌ的变量，但是在program语句块中是不行的
    
## AWK格式化
在将我们的AWK格式化命令的时候，我们还需要先来了解一下我们的Linux中的printf命令　　
### printf命令：
```
printf 'pattern' file1 file2 file3 ...
```
首先我们需要知道我们的printf命令会反复的调用我们的pattern去格式化我们的file文本串（贪婪的）直到文本串不够用为止  

格式替换符有：　　
    * \%s：字符串的形式输出我们的文本,默认是空串
    * \%f：浮点数形式输出我们的文本
    * \%b：转义输出我们的文本
    * \%c：字符
    * \%d：十进制，默认是0
    * \%o：无符号８进制
    * \%u：无符号十进制
    * \%x：无符号十六进制(a-f)
    * \%X：无符号十六进制(A-F)
    * \%\%:输出\%
其中我们需要注意一点就是这些格式替换福是可以采用格式修饰符修饰的，比如  
```
printf "%10s" lan    #默认是右对齐的１０个空位输出，不足补空格
printf "%-10s" lan    #默认是左对齐的１０个空位输出，不足的补空格
printf "%12.5f" 3.1415926535    #１２个空位精度是５，不足的补空格
printf "%10d" 123456      #默认的右对齐
printf "%-10d" 123456     #左对齐输出１０位，不足的补空格
printf "%+10d" 123456     #输出正负符号
printf "%+10d" -123456    #输出正负符号
```

### AWK内置的printf语句和print语句
和我们的上述描述的printf语句有几处不同
1. 在awk中我们的printf要使用,进行分割，否则就会无缝的拼接字符串
```
awk '{printf "%s\n%s\n","lantian","something"}'
```
2. 其中的替代的参数和我们的格式替换符的数目要必须一致，我们的awk内部的printf是不进行重复的替换的(和我们的shell下是不一样的)
3. 我们的内置变量也可以用于填充我们的格式替换符
4. 和我们的echo语句不一样，我们的内置的print,printf是不会默认输出换行符号的  
5. print和printf的区别
    * print和printf都支持转义，但是我们的print不支持格式替换
    * print支持我们的连续的输出，但是小心我们上述讨论的有无逗号的问题，但是我们的prinf更适合需要格式字符串的地方  
    
## AWK模式

我们可以简单的理解我们的模式其实就是我们的条件，我们每一次都是按照我们的一行来进行检索，所以我们的模式就是用来判断我们改行是不是满足条件，满足条件就执行后面对应的{}内的**动作组**　　

常用的关系运算符:<,<=,==,!=,>,>=,>,~,!~
其中的~表示的是正则的匹配，如果我们正则匹配成功的话，表达式就会返回True

模式的分类：
    * 空模式：我们如果没有制定模式的话，就是所有的行都匹配成功　
    * 特殊模式:BEGIN / END，这两个模式分别表示在处理文本前要执行的动作组，和处理完文本后要执行的动作组
    * 关系运算模式：利用我们的关系运算的性质，筛选符合条件的行执行相应的动作　
    ```
    awk '\$3=="root"{....}' file
    awk '\$3~/root/{...}' file
    awk '{if() action;else action}'
    awk '{for(i in 数组) action}'
    ```
